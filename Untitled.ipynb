```python
import time
import requests
import matplotlib as mp
import matplotlib.pyplot as plt

iss_api = "https://api.wheretheiss.at/v1/satellites/25544"
interval_req = 10 # pinged every 10s
points_req = 100 # 100 data points

# containers to update
latitudes = []
longitudes = []
timestamps = []

def ping_iss ():
    response = requests. get (iss_api, timeout=6)
    data = response.json ()
    latitude = float (data["latitude"])
    longitude = float (data["longitude"])
    timestamp = int (data["timestamp"])

    return latitude, longitude, timestamp

# sampling loop
for i in range (points_req):
    try:
        latitude, longitude, timestamp = ping_iss ()

        latitudes.append(latitude)
        longitudes.append(longitude)
        timestamps.append(timestamp)

    except Exception as ex:
        print("no request:", ex)

    if i < points_req - 1: # rest between samples, 10 seconds req
        time.sleep (interval_req)

# plotting
    plt.figure (figsize=(10,5))
    ax = plt.gca ()
 
    ax.set_title ("Tracking the ISS Path")
    ax.set_xlabel ("Longitude")
    ax.set_ylabel ("Latitude")

    ax.set_xlim (-180, 180) # degree spans of both
    ax.set_ylim (-90, 90)

    ax.grid (True, alpha = 0.2)

    ax.plot (longitudes, latitudes, linewidth = 1.2) # path readability

    ax.scatter (longitudes, latitudes, s=10, alpha = 0.6)

    plt.tight_layout() # tighten layout for preview

    plt.show()
```

```python
# Minimal ISS tracker: requests + matplotlib, 100 points @ 10s each

import time
import requests
import matplotlib.pyplot as plt

API = "https://api.wheretheiss.at/v1/satellites/25544"
INTERVAL_SECONDS = 10     # ping every 10 seconds (as required)
NUM_POINTS = 100          # collect at least 100 data points

# 1) Storage for the collected data
lats, lons, stamps = [], [], []   # stamps keeps the integer unix timestamps

def ping_iss():
    """
    Call the API once and return (lat, lon, ts).
    The API returns a FLAT JSON with keys: latitude, longitude, timestamp, etc.
    """
    r = requests.get(API, timeout=6)
    r.raise_for_status()
    d = r.json()
    return float(d["latitude"]), float(d["longitude"]), int(d["timestamp"])

# 2) Collect NUM_POINTS samples, spaced INTERVAL_SECONDS apart
for i in range(NUM_POINTS):
    try:
        lat, lon, ts = ping_iss()
        lats.append(lat); lons.append(lon); stamps.append(ts)
        print(f"{i+1:3d}/{NUM_POINTS}  t={ts}  lat={lat:.3f}  lon={lon:.3f}")
    except Exception as e:
        # If one request fails, just log and continue to the next interval
        print("request failed:", e)
    if i < NUM_POINTS - 1:
        time.sleep(INTERVAL_SECONDS)

# 3) Plot the path with matplotlib (simple lat/lon axes; no map libraries)
plt.figure(figsize=(10, 5))
ax = plt.gca()
ax.set_title("ISS Path (wheretheiss.at)")
ax.set_xlabel("Longitude (°)")
ax.set_ylabel("Latitude (°)")
ax.set_xlim(-180, 180)
ax.set_ylim(-90, 90)
ax.grid(True, alpha=0.3)

# draw the trail and mark start/end
ax.plot(lons, lats, linewidth=1)
ax.scatter(lons, lats, s=10, alpha=0.7)
ax.scatter([lons[0]],[lats[0]], s=60, marker="o", label="start")
ax.scatter([lons[-1]],[lats[-1]], s=60, marker="s", label="end")
ax.legend(loc="lower left")

plt.tight_layout()
plt.show()
```
